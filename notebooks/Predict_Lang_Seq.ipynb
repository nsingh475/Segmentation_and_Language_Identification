```python
import os
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
# spectrogram_1s_folder = 'RawData_img/test_1s_spectrogram/azz/'
language = 'na'
spectrogram_1s_folder = f'RawData_img/test_2s_spectrogram_4x/{language}/'
cnn_model_path = 'Models/slid_cnn_v1_e10.h5'
```

```python
from tensorflow.keras.models import load_model
model = load_model(cnn_model_path)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(rescale=1./255)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

def prepare_img(img_path):
    img = image.load_img(img_path, target_size=(128, 128))  # Specify target size
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array = img_array / 255.  # Normalize pixel values
    return img_array
```

```python
predicted_test_data = []
for spectrogram_folder in os.listdir(spectrogram_1s_folder):
    audio_name = spectrogram_folder
    predicted_lang_labels = []
    for spectrogram_img in os.listdir(spectrogram_1s_folder+spectrogram_folder):
        img_path = spectrogram_1s_folder+spectrogram_folder+'/'+spectrogram_img
        img = prepare_img(img_path)
        pred = model.predict(img)
        label = np.argmax(pred)
        predicted_lang_labels.append(label)
    
    predicted_test_data.append((audio_name, predicted_lang_labels))
```

```python
## Class Indexes
# {'ambos': 0, 'azz': 1, 'esp': 2, 'na': 3}
```

```python
import pandas as pd
df = pd.DataFrame(predicted_test_data, columns=['Audio Name', 'Language Labels'])
df.head(50)
```

```python
df.to_csv(f'Predictions/{language}_lang_labels_2s_4x_e10.csv', index=False)  # Set index=False if you don't want to include row numbers
print("DataFrame has been written to file.")
```
